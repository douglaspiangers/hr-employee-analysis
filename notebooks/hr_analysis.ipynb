## 1. Project Introduction

This project explores IBM HR employee data to understand workforce composition, attrition patterns, and the factors that influence employee turnover.

**Questions this analysis answers:**
- What is the overall attrition rate?
- Which departments have the highest turnover?
- How does monthly income vary across job roles?
- Are younger or lower-paid employees more likely to leave?
- What factors correlate most with income and tenure?

## 2. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
plt.rcParams['figure.dpi'] = 100
print('Libraries loaded successfully')
```

## 3. Load Dataset

```python
df = pd.read_csv('../data/ibm_hr_data.csv')

# Rename columns from Portuguese to English
df = df.rename(columns={
    'Idade': 'Age',
    'Evasão': 'Attrition',
    'Viagem_a_Trabalho': 'BusinessTravel',
    'Salário_Diário': 'DailyRate',
    'Departamento': 'Department',
    'Distância_de_Casa': 'DistanceFromHome',
    'Educação': 'Education',
    'Área_de_Educação': 'EducationField',
    'Número_do_Empregado': 'EmployeeNumber',
    'Satisfação_com_Ambiente': 'EnvironmentSatisfaction',
    'Gênero': 'Gender',
    'Salário_Horário': 'HourlyRate',
    'Envolvimento_no_Trabalho': 'JobInvolvement',
    'Nível_do_Cargo': 'JobLevel',
    'Cargo': 'JobRole',
    'Satisfação_no_Trabalho': 'JobSatisfaction',
    'Estado_Civil': 'MaritalStatus',
    'Salário_Mensal': 'MonthlyIncome',
    'Salário_Mensal_Taxa': 'MonthlyRate',
    'Nº_Empresas_Trabalhadas': 'NumCompaniesWorked',
    'Hora_Extra': 'OverTime',
    '%_Aumento_Salarial': 'PercentSalaryHike',
    'Avaliação_de_Desempenho': 'PerformanceRating',
    'Satisfação_com_Relacionamentos': 'RelationshipSatisfaction',
    'Nível_Opções_Ações': 'StockOptionLevel',
    'Total_Anos_Trabalhados': 'TotalWorkingYears',
    'Treinamentos_no_Ano': 'TrainingTimesLastYear',
    'Equilíbrio_Trabalho_Vida': 'WorkLifeBalance',
    'Anos_na_Empresa': 'YearsAtCompany',
    'Anos_no_Cargo_Atual': 'YearsInCurrentRole',
    'Anos_Desde_Última_Promoção': 'YearsSinceLastPromotion',
    'Anos_com_Gerente_Atual': 'YearsWithCurrManager',
})

print(f'Shape: {df.shape}')
df.head()
```

## 4. Data Cleaning

```python
# Check for missing values
print(f'Missing values: {df.isnull().sum().sum()}')

# Remove columns with only one unique value — no analytical use
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
print(f'Dropping constant columns: {constant_cols}')
df = df.drop(columns=constant_cols)

print(f'Final shape: {df.shape}')
print('Data cleaned successfully.')
```

## 5. Exploratory Analysis

```python
attrition_rate = (df['Attrition'] == 'Yes').mean() * 100

print('=== Workforce Overview ===')
print(f'Total employees:     {len(df):,}')
print(f'Average age:         {df["Age"].mean():.1f} years')
print(f'Avg monthly income:  ${df["MonthlyIncome"].mean():,.0f}')
print(f'Attrition rate:      {attrition_rate:.1f}%')
print(f'Departments:         {list(df["Department"].unique())}')
print()
df[['Age', 'MonthlyIncome', 'YearsAtCompany', 'TotalWorkingYears']].describe().round(2)
```

## 6. Data Visualizations

### 6.1 Employee Attrition Rate

This chart shows the proportion of employees who left the company versus those who stayed.

```python
attrition_count = df['Attrition'].value_counts().reset_index()
attrition_count.columns = ['Attrition', 'Count']
attrition_count['Percentage'] = (attrition_count['Count'] / len(df) * 100).round(1)

plt.figure(figsize=(6, 5))
colors = ['#2196F3', '#F44336']
bars = sns.barplot(data=attrition_count, x='Attrition', y='Count', palette=colors)
for i, row in attrition_count.iterrows():
    plt.text(i, row['Count'] + 5, f"{row['Percentage']}%", ha='center', fontsize=12, fontweight='bold')
plt.title('Employee Attrition Rate', fontsize=16, fontweight='bold')
plt.xlabel('Attrition', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.tight_layout()
plt.savefig('../images/attrition_rate.png', bbox_inches='tight')
plt.show()
```

### 6.2 Attrition by Department

Breaking down turnover by department shows which teams are most affected by employee exits.

```python
plt.figure(figsize=(9, 5))
sns.countplot(
    data=df, x='Department', hue='Attrition',
    palette={'Yes': '#F44336', 'No': '#2196F3'}
)
plt.title('Attrition by Department', fontsize=16, fontweight='bold')
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend(title='Left Company')
plt.tight_layout()
plt.savefig('../images/attrition_by_department.png', bbox_inches='tight')
plt.show()
```

### 6.3 Monthly Income Distribution

This histogram shows how salaries are spread — useful for identifying pay concentration and outliers.

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['MonthlyIncome'], bins=40, kde=True, color='steelblue')
plt.title('Monthly Income Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Monthly Income (USD)', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.tight_layout()
plt.savefig('../images/income_distribution.png', bbox_inches='tight')
plt.show()

print(f'Median income: ${df["MonthlyIncome"].median():,.0f}')
print(f'Average income: ${df["MonthlyIncome"].mean():,.0f}')
```

### 6.4 Monthly Income by Job Role

A boxplot comparing income ranges across roles — shows both median salary and how spread out each role's pay is.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='JobRole', y='MonthlyIncome', palette='Blues')
plt.title('Monthly Income by Job Role', fontsize=16, fontweight='bold')
plt.xlabel('Job Role', fontsize=12)
plt.ylabel('Monthly Income (USD)', fontsize=12)
plt.xticks(rotation=40, ha='right')
plt.tight_layout()
plt.savefig('../images/income_by_role.png', bbox_inches='tight')
plt.show()
```

### 6.5 Age vs Monthly Income — by Attrition

Each point is one employee. Color shows whether they left (red) or stayed (blue). Reveals which profile is more likely to churn.

```python
plt.figure(figsize=(9, 5))
sns.scatterplot(
    data=df, x='Age', y='MonthlyIncome', hue='Attrition',
    palette={'Yes': '#F44336', 'No': '#2196F3'}, alpha=0.5
)
plt.title('Age vs Monthly Income by Attrition', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Monthly Income (USD)', fontsize=12)
plt.legend(title='Left Company')
plt.tight_layout()
plt.savefig('../images/age_vs_income.png', bbox_inches='tight')
plt.show()
```

### 6.6 Correlation Heatmap

Shows which numeric variables are most related — useful for spotting patterns between age, income, satisfaction, and tenure.

```python
corr_cols = [
    'Age', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany',
    'JobSatisfaction', 'EnvironmentSatisfaction', 'WorkLifeBalance'
]
corr = df[corr_cols].corr()

plt.figure(figsize=(9, 7))
sns.heatmap(corr, annot=True, cmap='Blues', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix — Key HR Variables', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('../images/correlation_heatmap.png', bbox_inches='tight')
plt.show()
```

## 7. Key Insights

- The company has a **16.1% attrition rate** — about 1 in 6 employees left.
- **Sales** has the highest number of departures in absolute terms.
- Most employees earn between **$2,000 and $6,000/month**, with a long tail of high earners (Directors, Managers).
- **Younger, lower-paid employees** show a higher concentration of attrition (red dots in the bottom-left of the scatter plot).
- **Monthly income, total working years, and age** are strongly correlated — experience drives pay.
- **Satisfaction scores have a weak correlation with income**, suggesting pay alone does not determine how satisfied employees are.

## 8. Final Conclusion

This analysis of 1,470 IBM employees shows that attrition is concentrated among younger, lower-paid workers — particularly in Sales. The data points to early-career professionals as the highest-risk group for turnover.

Retention strategies focused on **competitive entry-level pay, career growth, and work-life balance** are likely to have the highest impact. The strong correlation between experience and income also highlights the long-term value of investing in employee development.
